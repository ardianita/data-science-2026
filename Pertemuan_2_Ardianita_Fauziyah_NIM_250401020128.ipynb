```python
import pandas as pd
import numpy as np

# Muat Dataset Titanic
URL = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(URL)
df.head()
```

```python
# Tipe Data Dataset Titanic
df.info()
```

```python
# Jumlah Kolom dan Baris Dataset Titanic
print(df.shape, df.columns.tolist())
print(f"Total Baris : {df.shape[0]}")
print(f"Total Kolom : {df.shape[1]}")
print(f"Kolom : {df.columns.tolist()}")
```

```python
# Kolom yang memiliki nilai kosong
print(df.isnull().sum())
```

Kolom yang memiliki nilai kosong : **Age**, **Cabin**, **Embarked**

```python
# Filter wanita kelas 1
wk1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]
print(f"Total Wanita Kelas 1 Selamat : {wk1['Survived'].sum()}")
print(f"Presentase Wanita Kelas 1 Selamat : {(wk1['Survived'].mean() * 100).round(2)} %")
```

```python
# Rata - rata usia wanita kelas 1
selamat = df[(df['Sex'] == 'female') & (df['Pclass']==1) & (df['Survived'] == 1)]
print(f"Rata - rata usia : {selamat['Age'].mean().round(1)} Tahun")
```

```python
# Kelas penumpang dengan tingkat keselamatan tertinggi
hasil = df.groupby('Pclass').agg(
    passenger=('PassengerId','count'),
    survived=('Survived','sum'),
    mean=('Survived','mean'),
    ).round(3)
hasil
```

Kelas dengan tingkat keselamatan tertinggi adalah **Kelas 1** dengan total penumpang yang selamat sebanyak **136 orang** atau setara dengan **63%** dari total penumpang.

```python
# Rata - rata tingkat keselamatan per kelas
persentase = df.groupby('Pclass')['Survived'].mean()
persentase.round(3)
```
